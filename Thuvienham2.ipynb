```python
import matplotlib.pyplot as plt
import numpy as np
def ham_bac_2(a,b,c,x):
  f = a*x*x + b*x + c
  return f
def ham_bac_3(a,b,c,d,x):
  f = a*x*x*x + b*x*x + c*x + d
  return f
def ham_bac_1(a,b,x):
  f = a*x + b
  return f
```

```python
fig, ax = plt.subplots()
x = np.linspace(start=-10.0, stop=10.0, num=200)
y2 = ham_bac_2(-2, -5, 3, x)
y1 = ham_bac_1(-5, 3, x)
y3 = ham_bac_3(2,-3,5,-9,x)
ax.plot(x, y1, label=r'$y=-5x+3$')
ax.plot(x, y2, label=r'$y=-2X^2-5x+3$')
ax.plot(x, y3, label=r'$y=2X^3-3x^2+5^x-9$')
ax.set_xlabel("Trục hoành - x")
ax.set_ylabel("Trục tung - y")
ax.legend()
plt.show()
```

```python
from matplotlib import cm
def ham_Rosenbrock(x, y):
  z = x*x + y*y - np.cos(12*x)- np.cos(18*y)
  return z
```

```python
x = np.linspace(start=-1, stop=1,
num=200)
y = np.linspace(start=-1.0, stop=1,
num=200)
x, y = np.meshgrid(x, y)
z = ham_Rosenbrock(x, y)
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
rosen_surf = ax.plot_surface(x, y, z, cmap=cm.gist_heat_r,
linewidth=0, antialiased=False)
ax.set_zlim(-2, 4)
fig.colorbar(rosen_surf, shrink=0.5, aspect=5)
plt.show()
```
