```python
import numpy as np
def ham_bac_2(a,b,c,x):
  f = a*x*x + b*x + c
  return f
```

1


```python
import matplotlib.pyplot as plt
```

2

```python
fig, ax = plt.subplots()
x = np.linspace(start=-10.0, stop=10.0, num=200)
y = ham_bac_2(2, 3, -5, x)
ax.plot(x, y, label=r'$y=2x^{2}+3x-5$')
ax.set_xlabel("Trục hoành - x")
ax.set_ylabel("Trục tung - y")
ax.plot(x, y)
label=r'$y=2x^{2}+3x-5$'
ax.set_title("Đồ thị phương trình bậc 2")
ax.legend()
plt.show()
```

```python
from matplotlib import cm
def ham_Rosenbrock(b, x, y):
  z = (x -1)**2 + b*(y - x**2)**2
  return z
```

2

```python
x = np.linspace(start=-2.0, stop=2.0,
num=200)
y = np.linspace(start=-1.0, stop=3.0,
num=200)
x, y = np.meshgrid(x, y)
z = ham_Rosenbrock(10, x, y)
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
rosen_surf = ax.plot_surface(x, y, z, cmap=cm.gist_heat_r,
linewidth=0, antialiased=False)
ax.set_zlim(0, 200)
fig.colorbar(rosen_surf, shrink=0.5, aspect=5)
plt.show()
```
