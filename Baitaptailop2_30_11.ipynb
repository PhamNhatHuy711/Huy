```python
n = abs(int(input()))
m = abs(int(input()))
y = np.random.randint(0,10,(m,n))
a = np.sqrt(y)
b = np.cos(y)
c = np.sin(y)
d = np.log(y)
e = np.tan(y)
print(np.round(a,3))
print(np.round(b,3))
print(np.round(c,3))
print(np.round(d,3))
print(np.round(e,3))
```
